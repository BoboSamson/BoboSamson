```python
import numpy as np
import matplotlib.pyplot as plt

def loss_function(theta):
    """Example quadratic loss function: L = (theta - 3)^2"""
    return (theta - 3) ** 2

def gradient(theta):
    """Derivative of the loss function: dL/dtheta = 2 * (theta - 3)"""
    return 2 * (theta - 3)

def momentum_gradient_descent(eta=0.1, beta=0.9, iterations=50):
    """Performs gradient descent with momentum."""
    theta = 0  # Initial value
    velocity = 0  # Initialize velocity
    theta_history = []
    
    for _ in range(iterations):
        grad = gradient(theta)
        velocity = beta * velocity - eta * grad  # Update velocity
        theta += velocity  # Update parameter
        theta_history.append(theta)
    
    return theta_history

def nesterov_accelerated_gradient(eta=0.1, beta=0.9, iterations=50):
    """Performs Nesterovs Accelerated Gradient descent."""
    theta = 0  # Initial value
    velocity = 0  # Initialize velocity
    theta_history = []
    
    for _ in range(iterations):
        lookahead_theta = theta + beta * velocity  # Compute lookahead position
        grad = gradient(lookahead_theta)
        velocity = beta * velocity - eta * grad  # Update velocity
        theta += velocity  # Update parameter
        theta_history.append(theta)
    
    return theta_history

# Run optimization methods
theta_momentum = momentum_gradient_descent()
theta_nesterov = nesterov_accelerated_gradient()

# Visualization
plt.figure(figsize=(8, 5))
plt.plot(theta_momentum, label='Momentum GD', linestyle='--', marker='o')
plt.plot(theta_nesterov, label='Nesterov GD', linestyle='-', marker='s')
plt.axhline(y=3, color='r', linestyle='dashed', label='Optimal Value (theta=3)')
plt.xlabel("Iterations")
plt.ylabel("Theta Value")
plt.title("Gradient Descent with Momentum vs. Nesterov")
plt.legend()
plt.show()
```
